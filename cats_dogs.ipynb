```python
label_list=[]

for item in new_data_folder:
    if item[:3]=="cat":
        label_list.append(1)
    else:
        label_list.append(0)

print(label_list)    
```

```python
import cv2
import glob

image_directory="C:\\Users\\haider\\Desktop\\New folder (3)\\kagglecatsanddogs_5340\\PetImages\\catvsdog\\"
image_extension=["png","jpg"]

files=[]
[files.extend(glob.glob(image_directory+"*."+e)) for e in image_extension]

dog_cat_images=np.asarray([cv2.imread(file) for file in files])
```

```python
dog_cat_images.shape
```

```python
dog_cat_images
label=np.asarray(label_list)
label
unique, counts = np.unique(label, return_counts=True)
print(counts,unique)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y_set, test_y_set=train_test_split(dog_cat_images,label, test_size=.2, random_state=42)
print(train_x.shape, test_x.shape, train_y_set.shape, test_y_set.shape)
```

```python
unique, counts = np.unique(train_y_set, return_counts=True)
counts
```

```python
train_x_set=train_x/255
test_x_set=test_x/255
```

```python
import tensorflow as tf
```

```python
mobilenet_model = tf.keras.applications.MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
mobilenet_model.trainable=False
```

```python
model=tf.keras.Sequential([
    mobilenet_model,
    tf.keras.layers.Flatten(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(2, activation="sigmoid")
])
model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss='sparse_categorical_crossentropy',
    metrics=["accuracy"]
)
```

```python
print(train_x_set.shape,train_y_set.shape)
train_y_set
```

```python
history=model.fit(train_x_set,train_y_set, epochs=5)
```

```python
# loss, accuracy = model.evaluate(train_x_set,train_y_set)
```

```python
loss, accuracy =model.evaluate(test_x_set,test_y_set)
```
